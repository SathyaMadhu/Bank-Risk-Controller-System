```python
def convert_to_yolo_format(x_min, y_min, x_max, y_max, img_width, img_height):
    x_center = (x_min + x_max) / 2.0 / img_width
    y_center = (y_min + y_max) / 2.0 / img_height
    width = (x_max - x_min) / img_width
    height = (y_max - y_min) / img_height
    return x_center, y_center, width, height
```

```python
for _, row in Data_tr.iterrows():
    img_filename = row['filename']
    class_id = row['class']
    x_min, y_min, x_max, y_max = row['xmin'], row['ymin'], row['xmax'], row['ymax']

    # Get image dimensions
    img_path = os.path.join(tr_data, img_filename)
    img = Image.open(img_path)
    img_width, img_height = img.size

    # Convert to YOLO format
    x_center, y_center, width, height = convert_to_yolo_format(x_min, y_min, x_max, y_max, img_width, img_height)

    # Create label file
    label_filename1 = os.path.splitext(img_filename)[0] + '.txt'
    label_path1 = os.path.join("data_yolo/images/labels", label_filename1)

    with open(label_path1, 'a') as f:
        f.write(f"{class_id} {x_center} {y_center} {width} {height}\n")

print("Labels converted to YOLO format.")
```
